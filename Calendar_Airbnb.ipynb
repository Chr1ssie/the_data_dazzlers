## This code will extract .gz file

```python
import gzip
import shutil
with gzip.open('extract.csv.gz', 'rb') as f_in:
    with open('calendar.csv', 'wb') as f_out:
        shutil.copyfileobj(f_in, f_out)
```

## This code cuts up csv to 15 chops

```python
import pandas as pd
```

```python
df=pd.read_csv("calendar.csv")
df
```

```python
import numpy as np
```

```python
df_list=np.array_split(df,15)
```

```python
for i in range(15):
    df_list[i].to_csv(f"./calendardata/calendar_{i+1}.csv",index=False)    
```
